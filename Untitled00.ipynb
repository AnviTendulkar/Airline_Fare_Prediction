```python
def date_time_preprocessig(df):
  df['Date_of_Journey'] = pd.to_datetime(df['Date_of_Journey'] ,format = '%d/%m/%Y')
  df['journey_day'] = df['Date_of_Journey'].dt.day
  df['journey_month'] = df['Date_of_Journey'].dt.month
  df['journey_year'] = df['Date_of_Journey'].dt.year
  df[['dep_hour' , 'dep_minute']] = df['Dep_Time'].str.split(':' ,expand=True)
  df.drop(columns=['Date_of_Journey','Dep_Time'] , inplace=True)
  print("ddd = ",df)
  return df
```

```python
def total_stops_preprocessor(df):
  df['Total_Stops'] = df['Total_Stops'].replace({'1 stop': 1, 'non-stop': 0, '2 stops': 2, '3 stops': 3, '4 stops': 4})
  return df
```

```python
def Arrival_Date_calculation(row):
    pattern = re.compile(r'(?:(\d+)h)?\s?(?:(\d+)m)?')
    match = pattern.match(row['Duration'])
    if match:
        hours = int(match.group(1)) if match.group(1) else 0
        minutes = int(match.group(2)) if match.group(2) else 0

    # Concatenate 'Date_of_Journey' and 'Dep_Time'
    row['Duration'] = (hours * 60 ) + minutes
    date_of_journey = row['Date_of_Journey'] + ' ' + row['Dep_Time']

    # Convert the concatenated string to datetime
    journey_date_time = datetime.strptime(date_of_journey, "%d/%m/%Y %H:%M")

    # Calculate arrival date by adding 'Duration_hour' and 'Duration_minute' to 'Date_of_Journey'
    arrival_date = journey_date_time + timedelta(hours=hours, minutes=minutes)
    return arrival_date
```

```python
class ArrivalTimeCalculator(BaseEstimator, TransformerMixin):
    def fit(self, df, y=None):
        return self

    def transform(self, df):
        date1 = df.apply(Arrival_Date_calculation ,axis=1)
        df['Arrival_Year'] = date1.dt.year
        df['Arrival_Month'] = date1.dt.month
        df['Arrival_Day'] = date1.dt.day
        df['Arrival_hour'] = date1.dt.hour
        df['Arrival_minute'] = date1.dt.minute

        print(type(df['Arrival_Year']),df.head(2))
        df.drop(columns=['Date_of_Journey','Dep_Time'] , inplace=True)
        return df

    def get_feature_names_out(self, input_features=None):
        # Return the output feature names after transformation
        # print("xxxx = ",df.columns)
        # return ['Date_of_Journey', 'Dep_Time']+['Arrival_Year' ,'Arrival_Month' ,'Arrival_Day','Duration']
        return ['Duration','Arrival_Year' ,'Arrival_Month' ,'Arrival_Day','Arrival_hour','Arrival_minute']


    # def get_params(self, deep=True):

    #     return {}
```

```python
preprocessing = ColumnTransformer([
    ('airline' ,ce.TargetEncoder(),['Airline']),
    ('one-hot', OneHotEncoder(drop='first', sparse_output=False, dtype='int16' ,handle_unknown = 'ignore'), [ 2, 3]) ,
    ('date_time_preprocessig' ,FunctionTransformer(date_time_preprocessig ,validate =False),['Date_of_Journey' ,'Dep_Time' ] ) ,
    ('ArrivalTimeCalculator', ArrivalTimeCalculator(), ['Date_of_Journey', 'Dep_Time','Duration']),
    ('total_stop',FunctionTransformer(total_stops_preprocessor,validate =False) , ['Total_Stops'])
] )
```
